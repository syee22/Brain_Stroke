## Modeling

```python
raw_data['stroke'].value_counts()
```

Analyzing the data we can see that we have a lot more data with Non Stroke, so the models will learning more about this data than when the person doesn't has Stroke and can't learn about when this person will has Brain Stroke, then we need to balance the classes and see what we can do with this data.

Here we use the function OverSampling to balancing the classes, we are doing a Oversampling and both target variable will have the same number of samples.

```python
raw_data
```

```python
X = raw_data.drop('stroke',axis=1).values
y = raw_data['stroke'].values
```

pip install imbalanced-learn

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=0)
X , y = ros.fit_resample(X,y)
```

```python
sns.countplot(x=y)
```

#### Label Encoder "범주형변수를 연속형변수로 변환"

```python
raw_data.info()
```

- 범주형변수 이름과 열 위치번호 추출

```python
category_feature = [ col for col in raw_data.columns if raw_data[col].dtypes == "object"]
category_feature
```

```python
index_num = [raw_data.columns.get_loc(f) for f in category_feature]
index_num
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder

X[:,0] = LabelEncoder().fit_transform(X[:,0])
X[:,4] = LabelEncoder().fit_transform(X[:,4])
X[:,5] = LabelEncoder().fit_transform(X[:,5])
X[:,6] = LabelEncoder().fit_transform(X[:,6])
X[:,9] = LabelEncoder().fit_transform(X[:,9])
```

```python
print(X)
```

```python
X_normalization[0]
```

```python
nb_scores= classification_report(y_test,nb_pred)
print(nb_scores)
```

# 결론

- Brain_Stroke 데이터에는 수치형변수와 범주형 변수들이 있으며 시각화를 통해 인사이트를 얻을 수 있다.
    - 뇌졸증의 위험율은 나이가 많을수록 올라갈 것이라 알 수 있다.
    - 나이에 의한 영향을 많이 받으므로 목표하는 변수들을 균등화하여 분석할 수 있도록 OverSampling을 이용해 샘플링을 더 크게 가졌다.
    - 위와 같은 내용을 한번 더 확인할 수 있는 분석으로 DecisionTree 와 RandomForest 모델에서 가장 중요한 feature로 나이 변수인 것이 확인되었다.
    - 뇌졸증과 같은 예민한 문제에서는 positive라고 분류한 것 중 실제 positive가 얼마나 되는지, 즉 실제 뇌졸증인 것이라 판단되었을 때 실제로 뇌졸중이라고 판단하는 데 사용되는 정밀도를 사용하는 것이 중요하다.
    - 따라서 Precision (정밀도)로 예측률을 평가하는 것이 적당하다.
    - 본 데이터에서는 precision이 가장 높게 나오는 것은 KNN 으로 예측률이 97% 나 되지만 이것은 과적합된 모델이라고 할 수 있다.
<br>

    - RBF 를 사용하는 SVM 모델과 Poly와 SandardScaler를 사용하는 SVM 이 가장 적합한 모델이다.
